```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
date_range = pd.date_range(start='1/1/2024', end='1/31/2024', freq='D')
date_range
```

```python
temp_data= np.random.normal(loc=10, scale=5, size=len(date_range))
temp_data
```

```python
temp_df= pd.DataFrame({'Date': date_range, 'Temperature': temp_data})
temp_df
```

```python
temp_df.set_index('Date', inplace=True)
```

```python
temp_df.head()
```

```python
plt.figure(figsize=(8,3))
plt.plot(temp_df.index, temp_df['Temperature'])
plt.title('Daily Temp in Jan24')
plt.xlabel('Date')
plt.ylabel('Temp in Celsius')
plt.show()
```

```python
plt.figure(figsize=(8,3))
pd.plotting.autocorrelation_plot(temp_df['Temperature'])
plt.title("Autocorrelation plot")
plt.show()
```

```python
plt.figure(figsize=(8,3))
plt.hist(temp_df['Temperature'], bins=10)
plt.title("Histogram for Daily Temp in Jan24")
plt.xlabel("Temperature in Celsius")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,3))
sns.kdeplot(temp_df['Temperature'], fill=True)
plt.title('Density plot for Daily Temp in Jan24')
plt.xlabel('Temperature')
plt.show()
```

```python
plt.figure(figsize=(2,2))
sns.boxplot(x=temp_df['Temperature'])
plt.title('Box and Whixker plot for Daily Temp in Jan24')
plt.xlabel('Temperature')
plt.show()
```

```python
plt.figure(figsize=(2,2))
sns.heatmap(temp_df, cmap='coolwarm')
plt.title('Heatmap for Daily Temp in Jan24')
plt.xlabel('Temperature')
plt.show()
```
