# EDA of The Bill Charge Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/bill charge.csv')
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.dtypes
```

```python
df['DayofWeek'] = df['Date'].dt.dayofweek  # Monday=0, Sunday=6
df['Month'] = df['Date'].dt.month
```

### Feature Selection

```python
df.columns
```

I am not selecting `date` feature

```python
features = ['DayofWeek', 'Month', 'Age', 'Patient Name', 'Bill Charge']
selected_df = df[features]
```

```python
selected_df.head()
```

### One Hot Encoding

```python
selected_df = pd.get_dummies(selected_df)
```
