# Decomposing a Time Series Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Daily Minimum Temperatures

```python
!wget https://raw.githubusercontent.com/jbrownlee/Datasets/refs/heads/master/daily-min-temperatures.csv
```

```python
df = pd.read_csv('/content/daily-min-temperatures.csv')

df.head()
```

```python
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(df['Temp'], model='additive', period=365)
fig = result.plot()
fig.set_size_inches(16, 9)
plt.show()
```
